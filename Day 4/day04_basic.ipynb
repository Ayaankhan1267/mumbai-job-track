```python
# Function without parameters
def say_hello():
    print("Hello, Ayaan!")

say_hello()

# Function with parameters
def greet(name):
    print(f"Hello, {name}!")

greet("Sara")
greet("Ali")

# Function with return value
def add(a, b):
    return a + b

print(add(5, 7))   # 12
```

```python
def greet(name="Guest"):
    print(f"Hello, {name}!")

greet()       # Hello, Guest!
greet("Ayaan")  # Hello, Ayaan!
```

```python
import csv

# Write CSV
with open("numbers.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["num1", "num2"])
    writer.writerow([10, 20])
    writer.writerow([30, 40])

# Read CSV
with open("numbers.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

```python
fruits = ["Apple", "Banana", "Mango", "Orange", "Grapes"]

with open("fruits.txt", "w") as f:
    for uit in fruits:
        f.write(uit + "\n")

with open("fruits.txt", "r") as f:
    print(f.read())
```

```python
import csv

FILENAME = "contacts.csv"

# Load contacts from file
def load_contacts():
    contacts = []
    try:
        with open(FILENAME, "r") as f:
            reader = csv.DictReader(f)
            contacts = list(reader)
    except FileNotFoundError:
        pass
    return contacts

# Save contacts to file
def save_contacts(contacts):
    with open(FILENAME, "w", newline="") as f:
        fieldnames = ["name", "phone", "email"]
        writer = csv.DictWriter(f, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(contacts)

# Add a contact
def add_contact(contacts, name, phone, email):
    contacts.append({"name": name, "phone": phone, "email": email})
    save_contacts(contacts)

# View contacts
def view_contacts(contacts):
    if not contacts:
        print("No contacts found.")
        return
    for c in contacts:
        print(f"{c['name']} | {c['phone']} | {c['email']}")

# Delete contact
def delete_contact(contacts, name):
    contacts = [c for c in contacts if c["name"] != name]
    save_contacts(contacts)
    return contacts

# Main program
def main():
    contacts = load_contacts()
    
    while True:
        print("\n--- Contact Book v2 ---")
        print("1. View Contacts")
        print("2. Add Contact")
        print("3. Delete Contact")
        print("4. Exit")
        
        choice = input("Enter choice: ")
        
        if choice == "1":
            view_contacts(contacts)
        elif choice == "2":
            name = input("Name: ")
            phone = input("Phone: ")
            email = input("Email: ")
            add_contact(contacts, name, phone, email)
        elif choice == "3":
            name = input("Enter name to delete: ")
            contacts = delete_contact(contacts, name)
        elif choice == "4":
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()    
```
